```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma

cv_data = [
    "Dati Anagrafici: Giuliana Pecchio, residente a Palerno, email: giulianapcch@libero.it, numero di telefono: +39 339 36 36 366"
    "Istruzione:2026/in corso – Certificazione di specializzazione in Agentic AI & Python presso Aulab,(BA)"
"2026 – Certificazione Web Application Developer Certipass presso Aulab,(BA)"
"2020 – Certificazione CIPD Livello 5 in Gestione delle Risorse Umane presso ICS Learn, (UK)"
"2018 – Certificazione CIPD Livello 3 in Pratiche di Risorse Umane presso ICS Learn, (UK)"
"Esperienza Lavorativa: GIU 2025 – SET 2025 | Receptionist Cooking Class - SI.LU.ANT. SRL – Amalfi (SA), Italia"
"Fornito supporto informativo multilingue su orari, logistica, trasporti pubblici e dettagli del territorio, garantendo un'esperienza utente eccellente."
"APR 2022 – DIC 2025 | Specialista del Servizio Clienti – Optima Italia S.p.A. – Napoli, Italia"
"Analisi dei dati statistici per determinare il livello del servizio clienti offerto."
"Competenze Tecniche:HTML5, CSS, JavaScript, PHP, Laravel, Python, Git, GitHub,Database SQL, Microsoft Office, CRM e Customer Management Tools"
"Competenze Trasversali:Adattamento, pianificazione e organizzazione, Gestione del tempo, Lavoro di squadra, Apprendimento continuo. "
"Lingue: Italiano: Madrelingua, Inglese: Avanzato, Spagnolo: Base. "

]

embeddings_model = OpenAIEmbeddings(model="text-embedding-3-small")

vector_db = Chroma.from_texts(
    texts=cv_data,
    embedding=embeddings_model
)

print(f" Inseriti con successo {len(cv_data)} blocchi di testo nel Vector DB!")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

q_recruiter = "Il candidato conosce framwork per lavorare con LLM (Large Language Model)?"

retriever = vector_db.as_retriever(search_kwargs={"k":2})
doc_retriever = retriever.invoke(q_recruiter)
context = "\n".join([doc.page_content for doc in doc_retriever])

template_prompt = ChatPromptTemplate.from_messages([
    ("system", "Sei un assistente HR. Rispondi alla domanda basandoti solo sul contest del cv fornito. Se non trovi l'informazione, dici chiaramente che non è presente."),
    ("human", "Contesto del CV:\n{context}\n\nDomanda:{question}")
])

prompt_finale = template_prompt.format_messages(context=context, question=q_recruiter)

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.2)
answer = llm.invoke(prompt_finale)

print("--- RISPOSTA AI RECRUITER ---")
print(answer.content)
```
