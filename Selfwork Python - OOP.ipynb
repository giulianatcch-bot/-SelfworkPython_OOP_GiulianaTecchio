 Esercizi:
Creare una classe Person , estenderla con due sottoclassi Dipendente e Studente
Estendere Dipendenti con 2 sottoclassi Docente e Freelance

Una classe deve essere astratta e ogni classe deve avere delle proprietà specializzanti

```python
from abc import ABC, abstractmethod

class Person(ABC):
    def __init__(self, name:str, surname:str, age:int):
        self.name = name
        self.surname = surname
        self.age = age

    @abstractmethod
    def show(self):
            pass
    def sayHello(self):
            return f"Hello, I'm {self.name} {self.surname}, I'm {self.age} years old."
        
class Student(Person):
    def __init__(self,name:str, surname:str, age:int, facolty:str):
        super().__init__(name,surname,age)
        self.facolty = facolty
    def show(self):
        return f"I'm a Student of {self.facolty}."

class Employee(Person):
    def __init__(self,name:str, surname:str, age:int, sector:str):
        super().__init__(name,surname,age)
        self.sector = sector
    def show(self):
        return f"I'm an employee by {self.sector} sector." 
    
class Teacher(Employee):
    def __init__(
        self, name:str, surname:str, age:int, sector:str, subject:str):
        super().__init__(name, surname, age, sector)
        self.subject = subject
    def show(self):
        return f"I'm a Teacher of {self.subject} into the {self.sector} sector."
    
class Freelancer(Employee):
    def __init__(
            self,
            name:str,
            surname:str,
            age:int,
            sector:str,
            vatNum:int,
    ):

            super().__init__(name, surname, age, sector)
            self.vatNum = vatNum
    def show(self):
        return f"I am a Frelancer in {self.sector} sector and my Vat Number is {self.vatNum}."

    
student1 = Student("Giuliana", "Tecchio", 30, "IT programming")
print(student1.sayHello())
print(student1.show())
print("-" * 30)

teacher1 = Teacher("Mario", "Rossi", 45, "public", "mathematics")
print(teacher1.sayHello())
print(teacher1.show())
print("-" * 30)

freelancer1 = Freelancer( 
    "Maria", "Rossi", 38, "wedding and cerimonies", 1564379
)
print(freelancer1.sayHello())
print(freelancer1.show())

                         


```
