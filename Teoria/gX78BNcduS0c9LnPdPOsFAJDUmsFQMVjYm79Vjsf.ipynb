# ChatCompletion API

Le API di OpenAI per il completamento del testo consentono di generare testo basato su un prompt iniziale. Gli utenti inviano una richiesta con un prompt e parametri opzionali per personalizzare il comportamento del modello, come la temperatura (che regola la creatività) e la lunghezza massima della risposta. L'API restituisce quindi una continuazione del testo, che può essere usata per vari scopi, dalla generazione di contenuti alla risposta a domande o al completamento di frasi.

[Documentazione Open AI](https://platform.openai.com/docs/guides/text-generation/quickstart)

[API Reference](https://platform.openai.com/docs/api-reference/chat)

[Open AI Models](https://platform.openai.com/docs/models)

[Open AI API Pricing](https://openai.com/api/pricing/)

[Open AI Platform](https://platform.openai.com/)

## Parametri


````python
completion = client.chat.completions.create(
    model="gpt-4.1",  # Modello da utilizzare, es. gpt-4o gpt-4.1 gpt-4.1-mini
    messages=[  # Lista dei messaggi nella conversazione
        {"role": "developer", "content": "Sei un assistente utile e cortese. usa massimo 1350 caratteri. Rispondi con un linguaggio forbito."},  # Messaggio di sistema
        {"role": "user", "content": "Raccontami della seconda guerra mondiale."}  # Messaggio dell'utente
    ],
    temperature=0,  # Valore tra 0 e 2. Più alto è, più creative sono le risposte.
    top_p=1.0,  # Alternativa alla temperatura. Definisce la probabilità cumulativa per il campionamento.
    n=1,  # Numero di risposte da generare per ogni messaggio.
    stream=False,  # Se True, la risposta viene restituita in streaming anziché tutta insieme.
    stop=None,  # Fino a 4 sequenze dove l'API smetterà di generare ulteriori token.
    max_tokens=2000,  # Limite massimo di token da generare nella risposta.
    presence_penalty=0.0,  # Penalità per scoraggiare il modello dal ripetere argomenti.
    frequency_penalty=0.0,  # Penalità per scoraggiare il modello dal ripetere le stesse parole.
)

print(completion.choices[0].message.content)
````

2. **messages**: La struttura della conversazione. Ogni messaggio ha un `role` e un `content`.
   - **role**:
     - `"system/developer"`: Definisce il contesto e il comportamento dell’assistente. Ad esempio, `"You are a helpful tutor."` orienta l'assistente a dare risposte educative e amichevoli.
     - `"user"`: Rappresenta i messaggi dell’utente che chiede informazioni o formula domande. Esempio: `{"role": "user", "content": "Tell me about photosynthesis."}`
     - `"assistant"`: Utilizzato per messaggi di risposta dall’assistente in contesti conversazionali prolungati.
   - 
     ```python
     messages=[
         {"role": "system", "content": "You are a knowledgeable science assistant."},
         {"role": "user", "content": "What is quantum mechanics?"},
         {"role": "assistant", "content": "Quantum mechanics is the branch of physics that studies subatomic particles."}
     ]
     ```

3. **temperature**: Controlla la creatività delle risposte. I valori variano tra **0 e 2**, con un valore **predefinito di 1**.
    - `temperature=0`: Produce risposte più coerenti e prevedibili.
   - `temperature=0.7`: Valore consigliato.
    - `temperature=1`: Bilancia creatività e coerenza.
    - `temperature=2`: Genera risposte creative ma meno prevedibili.

4. **top_p**: Limita le risposte a quelle più probabili. È un’alternativa alla `temperature`.
    - `top_p=0.9`: Il modello considera solo le risposte più probabili fino al 90% della probabilità cumulativa.

5. **n**: Specifica il numero di risposte da generare.
    - `n=2`: Riceverai due risposte diverse alla stessa domanda, utile per scegliere quella migliore.

6. **stream**: Se impostato su `True`, invia la risposta in tempo reale in blocchi parziali. Su `False` invia il testo completo una volta pronto.

7. **stop**: Definisce dove il modello dovrebbe fermarsi.
    - `stop=["\n"]`: Il modello si ferma quando rileva un carattere di nuova riga.
    - `stop=["Hello", "Ciao"]`: Blocca la risposta non appena raggiunge una delle stringhe indicate.

8. **max_tokens**: Limita la lunghezza della risposta generata dal modello.
    - `max_tokens=100`: La risposta sarà di massimo 100 token.

9. **presence_penalty** e **frequency_penalty**: Regolano la ripetitività del testo.
   - `presence_penalty`: Valori tra -2.0 e 2.0; un valore positivo incoraggia il modello a trattare argomenti nuovi, riducendo la tendenza a ripetere idee.
   - `frequency_penalty`: Valori tra -2.0 e 2.0; penalizza la ripetizione di frasi o parole già usate.

     - `presence_penalty=0.5` e `frequency_penalty=0.5` riducono la ripetizione di parole e incoraggiano il modello a coprire più concetti.


## Stream

````python
stream = client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=[
        {"role": "user", "content": "Ciao puoi aiutarmi a cucinare una cosa buona per cena? Suggeriscimi una ricetta leggera"}
    ],
    stream=True
)

# for chunk in stream:
#     print(chunk.choices[0].delta.content)

for chunk in stream:
    content = chunk.choices[0].delta.content
    if content:
        print(content, end='', flush=True)
        
# Per un effetto stampa in tempo reale:
# end='' impedisce che print() aggiunga automaticamente un newline.
# flush=True forza lo svuotamento del buffer, utile per vedere immediatamente l'output, specialmente in ambienti interattivi o script in tempo reale.
````

## Explain code

````python
completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "Vi verrà fornito un pezzo di codice e il vostro compito sarà quello di spiegarlo in modo conciso."},
        {"role": "user", "content": """
class Log:
    def __init__(self, path):
        dirname = os.path.dirname(path)
        os.makedirs(dirname, exist_ok=True)
        f = open(path, "a+")

        # Check that the file is newline-terminated
        size = os.path.getsize(path)
        if size > 0:
            f.seek(size - 1)
            end = f.read(1)
            if end != "\n":
                f.write("\n")
        self.f = f
        self.path = path

    def log(self, event):
        event["_event_id"] = str(uuid.uuid4())
        json.dump(event, self.f)
        self.f.write("\n")

    def state(self):
        state = {"complete": set(), "last": None}
        for line in open(self.path):
            event = json.loads(line)
            if event["type"] == "submit" and event["success"]:
                state["complete"].add(event["id"])
                state["last"] = event
        return state
        """}
    ]
)
````
